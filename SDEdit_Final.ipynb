# SDEdit — Colab Inference Notebook

This notebook is prepared for a GitHub repository and can be opened directly in Google Colab.

**What it does**
- Installs the required diffusion dependencies.
- Loads a Stable Diffusion image-to-image pipeline.
- Lets you upload and preview custom input images.
- Runs SDEdit-style image edits using configurable prompts and strength values.
- Saves generated outputs and an experiment log for report writing.

**Recommended runtime:** Google Colab with GPU enabled.

**How to use**
1. Open the notebook in Colab.
2. Run the cells from top to bottom.
3. Upload your test images in the upload cell.
4. Select a case in the case-selection cell.
5. Run the generation and logging cells.
